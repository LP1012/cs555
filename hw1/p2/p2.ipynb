## Problem 2 part b

plotting the growth factor G($\lambda \Delta t$) for the analytical solution, forward euler, backward euler, and CN

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ldts = np.linspace(-10, 0, 100)

analytical = np.exp(ldts)
forward_euler = (1 + ldts)
backward_euler = 1 / (1 - ldts)
CN = (1 + ldts / 2) / (1 - ldts / 2)
plt.figure(figsize=(10, 6))
plt.plot(ldts, analytical, label='Analytical', color='black')
plt.plot(ldts, forward_euler, label='Forward Euler', linestyle='--')
plt.plot(ldts, backward_euler, label='Backward Euler', linestyle='-.')
plt.plot(ldts, CN, label='Crank-Nicolson', linestyle=':')
plt.ylim(-0.5, 1.5)
plt.xlabel(r"G($\lambda \Delta t$)")
plt.ylabel(r"G($\lambda \Delta t$)")
plt.title(r"Growth Factor vs. $\lambda \Delta t$")
plt.legend()
plt.grid()
plt.savefig('growth_factors.png')
plt.show()
```
